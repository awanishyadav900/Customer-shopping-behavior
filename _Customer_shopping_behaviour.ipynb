```python
query = "SELECT * FROM cust"

data = pd.read_sql(query, engine)

data.head()
```

```python
# run sql query in jupyter
query = """
SELECT category,
 count(size) 
FROM cust
GROUP BY category
"""

result = pd.read_sql(query, engine)

result
```

```python
query = """ select category , sum(amounts) as total from cust group by category"""
result = pd.read_sql(query, engine)
result
```

```python
query = """ with cte as(
select customer_id,category, amounts,dense_rank() over(partition by category) as rnk from cust order by customer_id)
select * from cte where category = 'Clothing' """
result = pd.read_sql(query,engine)
result
```

```python
df122 = df.groupby('gender')['amounts'].sum().sort_values(ascending = False).reset_index()
```

```python
df122
```

```python
plt.figure(figsize=(3,3))
plt.pie(df122['amounts'],
       labels = df122['gender'],
       autopct= '%1.2f%%' )
plt.title("gender wise amounts")
plt.show()
```

```python
df123 = df.groupby('item_purchased').agg({'amounts':'sum', 'customer_id':'count'}).reset_index().sort_values("amounts" ,ascending= False)
```

```python
df123
```

```python
plt.figure(figsize=(15,5))
sns.barplot(df123, x = 'item_purchased', y = 'amounts')
plt.title("Total sales by items")
plt.xlabel("Item purchase")
plt.ylabel("Total amounts")
plt.show() 
```
